For the sake of completeness, let's analyze how the harmonic sum: $\sum_{k = 1}^n\frac{1}{k}$ can be approximated by $\log{n} + \gamma$, where $\gamma$ is the Euler-Mascheroni constant.

```python
# Define Euler-Mascheroni constant
GAMMA = 0.577215664901532
```

```python
harmonic_sum = 0
xs = []
harmonic_sums = []
approxs = []
errors = []
for i in range(1,100+1):
  harmonic_sum += 1/i
  approx = np.log(i) + GAMMA
  xs.append(i)
  harmonic_sums.append(harmonic_sum)
  approxs.append(approx)
  errors.append(harmonic_sum-approx)


plt.plot(xs,harmonic_sums,xs,approxs)
plt.show()
plt.plot(xs,errors)
plt.show()
```

We see that the two curves tend to overlap as $n$ grows, while the error reduced. Moreover, the approximation $\log{n} + \gamma$ underestimates the harmonic sum.
